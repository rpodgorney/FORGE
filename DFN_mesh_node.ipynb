All is good.  BUT...because the file is so big, it can't be edited in Excel, so need to add some additional columns to the file now, and then later add the 10 row header that is usually written by Leapfrog

Add the disp_k_bc	disp_j_bc	disp_i_bc	vertical	h_max	h_min columns

```python
data.insert(loc=8, column='disp_k_bc', value=1)
```

```python
data.insert(loc=9, column='disp_j_bc', value=1)
```

```python
data.insert(loc=10, column='disp_i_bc', value=1)
```

```python
data.insert(loc=11, column='vertical', value=1)
```

```python
data.insert(loc=12, column='h_max', value=1)
```

```python
data.insert(loc=13, column='h_min', value=1)
```

```python
data.rename(columns={'temperature_C': 'temperature'}, inplace=True)
```

```python
data.rename(columns={'pressure_MPa': 'pressure'}, inplace=True)
```

now check the file

```python
data.head()
```

now sort so values are in correct order for em2ex

```python
a = data.sort_values(["Z","Y","X"], ascending = [True, True, True])
```

```python
a.head()
```

now we have all the needed columns in the dataframe.  Now can populate the new columns with data, and also fix the pressure and temperature

```python
a['temperature'] =+ a['temperature'] + 273.15
```

```python
a['pressure'] = a['pressure']* 1e6
```

```python
a['disp_k_bc'] = (a['Sigma_V_MPa'] * -1e6)
```

```python
a['disp_j_bc'] = (a['sigma_h_max_MPa'] * -1e6)
```

```python
a['disp_i_bc'] = (a['sigma_h_min_MPa'] * -1e6)
```

to get IC for effective stress, must convert total to negtive and Pa, the add pressure.  Note that alpha (0.6) was already factored into the calculation of total stress, so we don't need to worry about it converting back into MOOSE conventions

```python
a['vertical'] = (a['Sigma_V_MPa'] * -1e6) + (a['pressure'])
```

```python
a['h_max'] = (a['sigma_h_max_MPa'] * -1e6) + (a['pressure'])
```

```python
a['h_min'] = (a['sigma_h_min_MPa'] * -1e6) + (a['pressure'])
```

```python
a.head()
```

now we can delete some columns

```python
del a['sigma_h_max_MPa']
```

```python
del a['sigma_h_min_MPa']
```

```python
del a['Sigma_V_MPa']
```

and now the dataframe should be ready for em2ex

```python
a.head()
```

now can write the data to a CSV file.

```python
a.to_csv("processed_n40e_node.csv", encoding='utf-8', index=False)
```
